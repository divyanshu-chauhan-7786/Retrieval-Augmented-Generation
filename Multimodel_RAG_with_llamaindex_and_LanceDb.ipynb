```python
from pytube import YouTube
def download_youtube_video(url, output_path):
    yt = YouTube(url)
    metadata = {"Author": yt.author, "Title": yt.title, "Views": yt.views}
    yt.streams.get_highest_resolution().download(output_path=output_path , filename="input_video.mp4")
    return metadata
```

```python
from moviepy.editor import VideoFileClip
import os
# video to image
def video_to_images(video_path, output_folder):
    clip = VideoFileClip(video_path)
    frame_dir = os.path.join(output_folder, "frame")
    os.makedirs(frame_dir, exist_ok=True)
    clip.write_images_sequence(
        os.path.join(frame_dir, "%04d.png"),
        fps=0.2
    )
    clip.close()
```

```python
def  video_to_audio(video_path, output_audio_path):
    clip = VideoFileClip(video_path)
    audio = clip.audio
    audio.write_audiofile(output_audio_path)
    clip.close()
```

```python
def audio_to_text(audio_path):
    r = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio_data = r.record(source)
        text = r.recognize_google(audio_data)
    return text
```

```python
from yt_dlp import YoutubeDL
import os

def download_youtube_video(url, output_path):
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
        'outtmpl': os.path.join(output_path, 'input_video.mp4'),
        'quiet': True,
        'noplaylist': True,
        'metadata': True,
        'js_runtimes': {'deno': {}}
    }
    with YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(url, download=False)
        if info_dict:
            metadata = {
                "Author": info_dict.get('uploader', 'unknown'),
                "Title": info_dict.get('title', 'unknown'),
                "Views": info_dict.get('view_count', 'unknown')
            }
        else:
            metadata = {"Author": "unknown", "Title": "unknown", "Views": "unknown"}
        ydl.download([url])
    return metadata
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings

# BAAI/bge-small-en-v1.5 is a good default
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
print("LlamaIndex is now configured to use HuggingFace embedding model: BAAI/bge-small-en-v1.5")
```

```python
docement  = SimpleDirectoryReader(output_folder).load_data()
```

```python
index = VectorStoreIndex.from_documents(docement, storage_context=Storage_context)
```

```python
retriver = index.as_retriever(similarity_top_k=1,image_similarity_top_k=3)
```

```python
from llama_index.core.response.notebook_utils import display_response, display_source_node
from llama_index.core.tools.query_engine import QueryEngineTool
from llama_index.core.schema import ImageNode
```

```python
def retrive(retriver,query_str):
  retriver_response = retriver.retrieve(query_str)
  retriver_image =[]
  retriver_text = []
  for doc in retriver_response:
    if isinstance(doc.node,ImageNode):
      retriver_image.append(doc.node.metadata["file_path"])
    else:
      display_source_node(doc, source_length = 200)
      retriver_text.append(doc.node.text)
  return retriver_image,retriver_text
```

```python
query_str = "what is the meaning of life?"
```

```python
retriver_image, retriver_text = retrive(retriver,query_str)
```

```python
from llama_index.multi_modal_llms.openai import OpenAIMultiModal
```

```python
open_multimodel = OpenAIMultiModal(model="gpt-4o-mini",api_key=OPENAI_API_KEY)
```

```python
qa_template_str = (
    "Context information is below.\n"
    "---------------------\n"
    "{context_str}\n"
    "---------------------\n"
    "Given the context information and not prior knowledge, answer the query.\n"
    "Query: {query_str}\n"
    "Answer: "
)
```

```python
result = open_multimodel.complete(qa_template_str.format(context_str=retriver_text,query_str=query_str),image_documents=retriver_image)
```
